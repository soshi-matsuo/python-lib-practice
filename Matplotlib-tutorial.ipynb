# MATLAB style & Objective Style

```python
import matplotlib.pyplot as plt
import matplotlib.style

matplotlib.style.use('ggplot')
```

```python
x = [1,2,3]
y = [2,4,9]

plt.plot(x,y)
plt.title('MATLAB-style')
plt.show()
```

```python
x = [1,2,3]
y = [2,4,9]

#fig = 描画オブジェクト、ax=サブプロット
fig, ax = plt.subplots()

ax.plot(x,y)
ax.set_title('OO_style')

plt.show()
```

```python
#O-O style

# set subplot as 1*1
fig,ax = plt.subplots()

#set axis labels
ax.set_xlabel('x label')
ax.set_ylabel('y label')

#set legend graph
ax.plot([1,2,3],[2,4,9],label='legend label')
ax.legend(loc='best')

plt.show()
```

```python
# plot 2 graphs
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [3,1,2]
ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```

```python
import numpy as np
x = np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig,ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
ax.legend()

plt.show()
```
